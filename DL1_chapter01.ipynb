```
#1.4. Python Script File
#1.4.1. 파일로 저장하기 -- 생략
#1.4.2. 클래스
class Man:
  def __init__(self, name): # 생성자, 첫번째 인수로 자신의 인스턴스를 나타내는 self를 쓰는 것이 명시적인 룰
    self.name = name
    print("Initialized!")
  def hello(self):
    print("Hello " + self.name + "!")
  def goodbye(self):
    print("Good-bye " + self.name + "!")

m = Man("David")
m.hello()
m.goodbye()
```

```
#1.5.3. Numpy에서의 산술 연산
x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])
print(x + y)
print(x - y)
print(x * y)
print(x / y)

# 스칼라값의 조합으로 산술된 연산 수행(Broadcast)
x = np.array([1.0, 2.0, 3.0])
x / 2.0
```

```
X = np.array([[51,55],[14,19],[0,4]])
print(X)
print(X[0])
print(X[0][1])
```

```
print(X > 15) # Bool
print(X[X>15]) # Value
```

```
#1.6. Matplotlib, 그래프 시각화
#1.6.1. 단순한 그래프 그리기
import matplotlib.pyplot as plt

# Data 준비
x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1 간격으로 생성
y = np.sin(x)

# Draw an graph
plt.plot(x,y)
plt.show()
```

```
#1.6.2. pyplot

# Data 준비
x = np.arange(0,6,0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# Draw graph
plt.plot(x, y1, label = "sin")
plt.plot(x, y2, linestyle="--", label = "cos") # cos, dot-line
plt.xlabel("x") # name of X axis
plt.ylabel("y") # name of Y axis
plt.title('sin & cos') # Title
plt.legend
plt.show()
```
